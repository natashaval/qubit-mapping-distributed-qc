```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Sample JSON structure
json_data = {
    "10": {
        "graphstate": {
            "init": { "size": 150, "depth": 26, "interval": 0.01913905143737793 },
            "full_10_2": {
                "basic_size": 189,
                "basic_depth": 59,
                "basic_interval": 0.09852337837219238,
                "basic_swap": 39,
                "sabre_size": 156,
                "sabre_depth": 33,
                "sabre_interval": 0.09065675735473633,
                "sabre_swap": 6,
                "lookahead_size": 158,
                "lookahead_depth": 33,
                "lookahead_interval": 0.10979127883911133,
                "lookahead_swap": 24
            },
            "line_5_4": {
                "basic_size": 390,
                "basic_depth": 110,
                "basic_interval": 0.30722975730895996,
                "basic_swap": 240,
                "sabre_size": 174,
                "sabre_depth": 28,
                "sabre_interval": 0.269437313079834,
                "sabre_swap": 24,
                "lookahead_size": 199,
                "lookahead_depth": 39,
                "lookahead_interval": 0.3493330478668213,
                "lookahead_swap": 147
            }
        }
    }
}

# Extract graphstate data
graphstate = json_data["10"]["graphstate"]

# Flatten the data into a list of dictionaries
data = []

for key, value in graphstate.items():
    if key == 'init':
        continue  # Skip the 'init' data as it doesn't have the same structure
    data.append({
        "graphstate": key,
        "type": "basic",
        "size": value.get("basic_size"),
        "depth": value.get("basic_depth"),
        "interval": value.get("basic_interval"),
        "swap": value.get("basic_swap")
    })
    data.append({
        "graphstate": key,
        "type": "sabre",
        "size": value.get("sabre_size"),
        "depth": value.get("sabre_depth"),
        "interval": value.get("sabre_interval"),
        "swap": value.get("sabre_swap")
    })
    data.append({
        "graphstate": key,
        "type": "lookahead",
        "size": value.get("lookahead_size"),
        "depth": value.get("lookahead_depth"),
        "interval": value.get("lookahead_interval"),
        "swap": value.get("lookahead_swap")
    })

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame to ensure it's correct
print(df)
```

```python
# Plotting the 'size' metric as a bar chart
plt.figure(figsize=(10, 6))

# Use the groupby method to get the correct plotting
df.groupby(['graphstate', 'type'])['size'].sum().unstack().plot(kind='bar', ax=plt.gca())

plt.title('Size by Graphstate and Type')
plt.ylabel('Size')
plt.xlabel('Graphstate')
plt.xticks(rotation=0)
plt.legend(title='Type')
plt.show()

# Similarly, you can plot 'depth', 'interval', and 'swap'
plt.figure(figsize=(10, 6))
df.groupby(['graphstate', 'type'])['depth'].sum().unstack().plot(kind='bar', ax=plt.gca())
plt.title('Depth by Graphstate and Type')
plt.ylabel('Depth')
plt.xlabel('Graphstate')
plt.xticks(rotation=0)
plt.legend(title='Type')
plt.show()

# Plotting the 'interval' metric
plt.figure(figsize=(10, 6))
df.groupby(['graphstate', 'type'])['interval'].sum().unstack().plot(kind='bar', ax=plt.gca())
plt.title('Interval by Graphstate and Type')
plt.ylabel('Interval (seconds)')
plt.xlabel('Graphstate')
plt.xticks(rotation=0)
plt.legend(title='Type')
plt.show()

# Plotting the 'swap' metric
plt.figure(figsize=(10, 6))
df.groupby(['graphstate', 'type'])['swap'].sum().unstack().plot(kind='bar', ax=plt.gca())
plt.title('Swap by Graphstate and Type')
plt.ylabel('Swap')
plt.xlabel('Graphstate')
plt.xticks(rotation=0)
plt.legend(title='Type')
plt.show()
```

```python
# Function to create subplots for a specific metric across all groups
def plot_metric_subplots(df, metric):
    groups = df['group'].unique()
    num_groups = len(groups)
    
    fig, axes = plt.subplots(nrows=num_groups, ncols=1, figsize=(14, 5 * num_groups))
    
    if num_groups == 1:
        axes = [axes]  # To make axes iterable when there's only one group
    
    for ax, group in zip(axes, groups):
        group_df = df[df['group'] == group]
        group_df.groupby(['graphstate', 'type'])[metric].sum().unstack().plot(kind='bar', ax=ax)
        ax.set_title(f'{metric.capitalize()} - {group.upper()}')
        ax.set_ylabel(metric.capitalize())
        ax.set_xlabel('Graphstate')
        ax.legend(title='Type')
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    
    plt.tight_layout()
    plt.show()

# Plotting the metrics
plot_metric_subplots(df, 'size')
plot_metric_subplots(df, 'depth')
plot_metric_subplots(df, 'interval')
plot_metric_subplots(df, 'swap')
```
